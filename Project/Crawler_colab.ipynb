# 2. Tải tất cả đường link và lọc ra những đường link chỉ chứa html:

```python
vnex_paper = newspaper.build('https://vnexpress.net/', language = 'vi', memoize_articles = False, follow_meta_refresh = True, keep_article_html=True)
```

```python
size = vnex_paper.size()
size
```

```python
valid_url = np.array([],np.string_)
for i in range(size):
  vnex_url = vnex_paper.article_urls()[i]
  # fillter out only-html link
  temp = ''
  for j in range(len(vnex_url)):
    temp = vnex_url[len(vnex_url) - j - 1] + temp
    if temp == 'html':
      valid_url = np.append(valid_url,vnex_url)
print(np.shape(valid_url)[0])
```

```python
star = np.array([],np.string_)
for i in range(np.shape(valid_url)[0]):
  vnex_url = Atc(valid_url[i],language = 'vi')
  vnex_url.download()
  vnex_url.parse()
  print(vnex_url.meta_data['tt_list_folder_name'])
```
